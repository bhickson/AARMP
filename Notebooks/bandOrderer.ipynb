```python
import os
import rasterio as rio
import glob
import numpy as np
```

```python
image_file = r"M:\Data\TrainingImageStack\m_3511406_se_11_1_20150528_TrainingStack.tif"
```

```python
band_columns = []
with rio.open(image_file) as ras_stack:
    for i in range(1, ras_stack.count + 1):
        column = ras_stack.tags(i)['NAME']
        band_columns.append(column)
    print(ras_stack.read().shape)
```

```python
all_files = glob.glob(r"M:\Data\TrainingImageStack" + "\\*.tif")
```

```python
count = 0
for file in all_files:
    bands_order = []
    fpath = os.path.join(r"M:\Data\TrainingImageStack", file)
    with rio.open(fpath) as ras_stack:
        prof = ras_stack.profile
        for i in range(1, ras_stack.count + 1):
            column = ras_stack.tags(i)
            
            bands_order.append(column)
            #bands_dict[column] = i
            
        if bands_order != band_columns:
            count += 1
            print(str(count) + " : " + fpath)
            write_over = True 
            bands_dict = {}
            for i in range(1, ras_stack.count + 1):
                #print(ras_stack.tags(i)['NAME'])
                
                #for i range(1, bands_order + 1):
                bands_dict[ras_stack.tags(i)['NAME']] = ras_stack.read(i)
            
            print(len(bands_dict))
            bands_list = []
            for name in band_columns:
                bands_list.append(bands_dict[name])
                
    if write_over == True:
        out_array_stack_np = np.stack(bands_list, axis=0)
        print(out_array_stack_np.shape)
        # print(out_array_stack_np.dtype)

        

        with rio.open(fpath, 'w', **prof) as outras:
            for n, tag in enumerate(band_columns):
                outras.update_tags(n+1, NAME=tag)
            outras.write(out_array_stack_np.astype(rio.int16))             
```
